```python
import pandas as pd
import time,datetime
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
data=pd.read_json("./data/DXYArea.json")
```

```python
pd.set_option("max_colwidth",300)
pd.set_option('display.max_columns', 1000)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', 1000)
```

```python
data.shape
```

```python
data[["continentName","countryName","provinceShortName","currentConfirmedCount","confirmedCount","suspectedCount",
      "curedCount","deadCount","updateTime"]]
```

```python
dataNew=data[["continentName","countryName","provinceName","currentConfirmedCount","confirmedCount","suspectedCount","curedCount",
              "deadCount","updateTime"]]
```

```python
dataNew["updateTime"] = pd.to_datetime(dataNew["updateTime"], unit='ms').dt.strftime('%Y-%m-%d, %H:%M:%S')
```

```python
dataNew["updateTime"] = pd.to_datetime(dataNew["updateTime"])
```

```python
dataNew.drop_duplicates('provinceName',keep='first',inplace=True)
```

```python
dataNew
```

```python
#各大洲疫情分布比例 累积确诊
pie=dataNew.groupby("continentName")["confirmedCount"].sum()
```

```python
pie
```

```python
pie.index
```

```python
pie.values
```

```python
plt.pie(
    pie.values,
    labels=pie.index,
    autopct='%.1f%%',
    pctdistance=0.8,)
plt.show()
```

```python
#累计确诊top10
bar=dataNew.sort_values(by="confirmedCount",ascending=False).head(10)
```

```python
bar
```

```python
bar["countryName"].tolist()
```
