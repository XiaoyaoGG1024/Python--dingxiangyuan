# 读取数据

```python
data=pd.read_json("./data/DXYArea.json")
```

```python
data["updateTime"] = pd.to_datetime(data["updateTime"], unit='ms').dt.strftime('%Y-%m-%d')
```

```python
data["updateTime"] = pd.to_datetime(data["updateTime"])
```

```python
data.dtypes
```

```python
data.head(1)
```

```python
dataNew=data[["continentName","countryName","provinceName","currentConfirmedCount","confirmedCount","suspectedCount","curedCount",
              "deadCount","updateTime"]]
dataNew
```

# 数据异常处理

```python
#keep='first'表示保留第一次出现的重复行
dataNew.drop_duplicates(subset=['provinceName','updateTime'],keep='first',inplace=True)
```

```python
dataNew.shape
```

```python
#缺失值处理
dataNew=dataNew.dropna()
```

```python
dataNew
```

```python
#查看数据的完整性
dataNew.isnull().sum()
```
